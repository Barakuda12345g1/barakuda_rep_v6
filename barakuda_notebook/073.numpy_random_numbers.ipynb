# Случайные числа в Numpy

### Генерация float

#### .random.rand()

```python
# получаем достп к библиотеке Numpy
import numpy as np
# получаем случайное число от 0 до 1
print(np.random.rand())

# разделительная черта нужна для читабельности кода
print('*' * 60)

# интересный способ получения случайного числа от 0 до 100
print(np.random.rand() * 100)

print('*' * 60)

# генерирую случайные числа в виде одномерного масива
# с заданным числом элементов
print(np.random.rand(5))

print('*' * 60)

# создание двумерного масива
print(np.random.rand(2, 3))

print('*' * 60)

# смотрим на то как выглядит многомерный масив
np.random.rand(2, 3, 3, 4, 5)
```

#### .random.sample()

```python
# передаем кортеж в функцию
shape = (2, 3)
# запоминаем, что кортеж не может быть предан в виде аргумента
np.random.rand(shape)
```

```python
# распаковка позволяет передать кортеж в аргумент
print(np.random.rand(*shape))

# раздельительная черта из звездочек дужна чтобы быстро читать и понимаьт код
print('*' * 60)

# находим результат и получаем способность внести в качестве аргумента 
# кортеж
np.random.sample(shape)
```

#### .random.uniform()

```python
# такая запись даст нам одно число в диапозоне от 0 до 1
print(np.random.uniform())

# разделяем вывод для лучшей читабельности
print('*' * 60)

# задаем новые ганицы диапозона
print(np.random.uniform(-30, 50))

print('*' * 60)

# изучаем метод, который позволяет устанавливать верхнюю границу,
# нижнюю границу, количество элементов в массиве
print(np.random.uniform(0.5, 0.75, size=5))

print('*' * 60)

# в аргументации можно указать размерность в виде кортежа
print(np.random.uniform(-1000, 500, size=(2, 3)))
```

## Генерация int

### .random.randint()

```python
# такая запись будет давать случайные числа от нуля до аргумента - 1
# и это при условии, что верхняя граница не указанна
print(np.random.randint(4, size=(2, 3)))

print('*' * 60)

# при указании и нижней и верхней границы, случайные числа будут
# выбираться из этого диапозона
print(np.random.randint(6, 12, size=(3, 3)))
```

## Генерация выборок

### .random.shuffle()

```python
# создаем одномерный массив
arr = np.arange(6)
# перемешиваем одномерный массив с помощью метода .shuffle(), 
# библиотеки .random
np.random.shuffle(arr)
arr
```

### .random.permutation()

```python
playlist = ["The Beatles", "Pink Floyd", "ACDC", "Deep Purple"]
# в отличии от .shuffle перемешанный масив может поместиться в обьекте
shuffeled = np.random.permutation(playlist)
# отмечаем, что основной список не изменился
print(playlist)
print(shuffeled)

# одновременно создаем массив из чесел, идущих по возастанию,
# и перемешиваем его 
np.random.permutation(10)
```

### .random.chooice()

```python
workers = ['Ivan', 'Nikita', 'Maria', 'John', 'Kate']

# сохраняем  в переменную заданное количество случайных чисел
choice = np.random.choice(workers, size= 2, replace=False)
print(choice)

# если в аргументе длинна больше чем список, возникнет ощибка
choice = np.random.choice(workers, size=10, replace=False)
choice
```

## seed генератора случайных чисел

### .random.seed(<np.uint32>)

```python
# задем значение seed к котрому будет привязана псевдослучайная
# генерация цифр
np.random.seed(23)
np.random.randint(10, size=(3, 4))
```

```python
# смотрим на повторимость случайной генерации цифр
np.random.seed(100)
print(np.random.randint(10, size=3))
print(np.random.randint(10, size=3))
print(np.random.randint(10, size=3))
```

## Задание 9.6

```python
# используем библиотеку Numpy
import numpy as np
# работаем с генерациями при seed равным 2021
np.random.seed(2021)
# пишем быстрый способ напечатать случайное число от 0 до 1
simple = np.random.rand()
# находим 120 случайных чисел в заданном диапозоне
randoms = np.random.uniform(-150, 2021, size=120)
# получаем целые случайные числа
table = np.random.randint(1, 101, size=(3, 2))
# использую быстрый способ создания последовательности чисел
even = np.arange(17)[::2]
# убираем первый элемент
even = even[1::]
# закрепляем знания о работе с функцией .copy()
mix = even.copy()
# закрепляем знания о работе метода .shuffle()
np.random.shuffle(mix)
# закрепляем знания о работе функции .choice()
select = np.random.choice(even, size=3, replace=False)
# в отличии от .shuffle(), .permutation() может быть возвращена
triplet = np.random.permutation(select)
print(simple, randoms, table, even, mix, select, triplet, sep='\n')
```
