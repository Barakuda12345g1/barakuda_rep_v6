## Задание 2.3

Напишите функцию delete_columns(df, col=[]), которая удаляет столбцы из DataFrame и возвращает новую таблицу. Если одного из указанных столбцов не существует в таблице, то функция должна возвращать None. 

```python
# импортирую библиотеку
import pandas as pd


customer_df = pd.DataFrame({
            'number': [0, 1, 2, 3, 4],
            'cust_id': [128, 1201, 9832, 4392, 7472],
            'cust_age': [13, 21, 19, 21, 60],
            'cust_sale': [0, 0, 0.2, 0.15, 0.3],
            'cust_year_birth': [2008, 2000, 2002, 2000, 1961],
            'cust_order': [1400, 14142, 900, 1240, 8430]
        })

# инициализируем функцию
def delete_columns(df, col=[]):
    # циклом работаем с каждым элементом отдельно
    for elem in col:
        # создаю условие проверяющее на наличие элемента в таблице
        if elem in df:
            df.drop([elem], axis=1, inplace=True)
        else:
            # выводим None если такого элемента не существует
            # иначе удаляем столбцы и возвращаем правленнцю верстию таблицы
            return None
            
    return df

display(delete_columns(customer_df, col=['number']))
```

## Задание 2.4

Задан DataFrame countries_df, содержащий следующие столбцы: название страны, население (population) в миллионах человек и площадь страны (area) в квадратных километрах.


Для каждой страны рассчитайте плотность населения (количество человек на квадратный километр).

Затем по полученным данным рассчитайте среднее по плотностям населения в указанных странах. Ответ округлите до сотых.

```python
countries_df = pd.DataFrame({
    'country': ['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'],
    'population': [56.29, 38.05, 322.28, 146.24, 45.5, 9.5, 17.04],
    'area': [133396, 9984670, 9826630, 17125191, 603628, 207600, 2724902]
})
# для правельного расчета плотности конвертируем данные

# провожу расчет в отеледльном ряду
density = countries_df['population'] / countries_df['area']
# привожу в нужный нам формат
density = density * 1000000
# округляем и воспроихводим
round(density.mean(), 2)
# создаю признак по плотности населения
countries_df['density'] = density
display(countries_df)
```

## Задание 3.3

Создайте в таблице melb_df признак WeekdaySale (день недели). Найдите, сколько объектов недвижимости было продано в выходные (суббота и воскресенье), результат занесите в переменную weekend_count. В качестве ответа введите результат вывода переменной weekend_count.

```python
# импортируем библиотеку
import pandas as pd


# читаем csv-файл
melb_data = pd.read_csv('data/melb_data_ps.csv', sep=',')
# оставляем исходную таблицу без изменени 
melb_df = melb_data.copy()

# приводим к формату datetime
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)
# создаем новый признак с днями недели
melb_df['WeekdaySale'] = melb_df['Date'].dt.dayofweek
# создаю маску для посчета через .shape
mask_one = melb_df['WeekdaySale'] == 6
mask_two = melb_df['WeekdaySale'] == 5
# помещаю ответ в переменную
weekend_count = melb_df[mask_one | mask_two].shape[0]
print(weekend_count)

# смотрим на интерисующий признак в формате Series
display(melb_df['WeekdaySale'])
```

## Задание 3.4 & 3.5

Вам представлены данные (в формате csv) об отчётах очевидцев НЛО в США за период с 1930 по 2020 год.

В данных есть следующие признаки:

        "City" — город, где был замечен НЛО;
        "Colors Reported" — цвет объекта;
        "Shape Reported" — форма объекта;
        "State" — обозначение штата;
        "Time" — время, когда был замечен НЛО (данные отсортированы от старых наблюдений к новым). 

Прочитайте данные, сделайте преобразование времени к формату datetime и выполните задания ниже.

```python
import pandas as pd


# читаем файл в переменную
nlo_data = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv', sep=',')

# копируем для работы
nlo_df = nlo_data.copy()
display(nlo_df.head(7))
# приводим прирзнак Time к фомрату datetime
nlo_df['Time'] = pd.to_datetime(nlo_df['Time'])

display(nlo_df)
```

В каком году отмечается наибольшее количество случаев наблюдения НЛО в США?

```python
# создадим признак с только с годами
nlo_df['Year'] = nlo_df['Time'].dt.year
display(nlo_df)
# смотрим на метод работы описательной статистики
display(nlo_df['Year'].value_counts())
```

Найдите средний интервал времени (в днях) между двумя последовательными случаями наблюдения НЛО в штате Невада (NV).

```python
# будем брать время в днях из нового признака 
nlo_df['Date'] = nlo_df['Time'].dt.date
# смотрим на таблицу
display(nlo_df[nlo_df['State'] == 'NV'])
# агрегиющий метод diff создает обьект типа Series, в которых указана 
# информация об интервалах между двумя последовательными значеними в столбце 
delta_date = nlo_df[nlo_df['State'] == 'NV']['Date'].diff()
display(delta_date)
# узнаем что среднее значение, приведенное в числовой формат
delta_date.dt.days.mean()
```
