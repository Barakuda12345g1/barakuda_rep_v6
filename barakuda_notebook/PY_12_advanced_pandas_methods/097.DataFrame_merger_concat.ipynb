## .concat([df1, df2,...], axis=)



        objs — список объектов DataFrame ([df1, df2,…]), которые должны быть сконкатенированы;
        axis — ось определяет направление конкатенации: 0 — конкатенация по строкам (по умолчанию), 1 — конкатенация по столбцам;
        join — либо inner (пересечение), либо outer (объединение);
        ignore_index — по умолчанию установлено значение False, которое позволяет значениям индекса оставаться такими, какими они были в исходных данных. Если установлено значение True, параметр будет игнорировать исходные значения и повторно назначать значения индекса в последовательном порядке.



```python
# читаю библиотеку
import pandas as pd


# знакомлюсь с новыми таблицами
raitings1_df = pd.read_csv('data/ratings1.csv', sep=',')
raitings2_df = pd.read_csv('data/ratings2.csv', sep=',')
movies_df = pd.read_csv('data/movies.csv', sep=',')
dates_df = pd.read_csv('data/dates.csv', sep=',')
```

```python
# добавлем вниз строки из второй таблицы в первую 
raitings = pd.concat([raitings1_df, raitings2_df])
display(raitings)
```

```python
# выстравляем индексы верно при помощи параметра  ignore_index=True
raitings = pd.concat(
    [raitings1_df, raitings2_df],
    ignore_index=True
)
display(raitings)
```

### .drop_duplicates(ignore_index=True)

```python
# смотрмин на количество строк в сконкатенированной раннее таблице
print(f'Количество строк в таблицу raitings: {raitings.shape[0]}')
# узнаем количество строк в таблице dates
print(f'Количество строк в таблице date_df: {dates_df.shape[0]}')
# узнаем что количество строк не совпадает
print(raitings.shape[0] == dates_df.shape[0])
# узнаем что есть дубликат в конце первой таблицы и в начале новой
display(raitings1_df.tail(1))
display(raitings2_df.head(1))
# убираем дубликаты и правим индекс
raitings.drop_duplicates(ignore_index=True)
print(f'Число строк в таблице raitings: {raitings.shape[0]}')
# конкатенируем таблицы по столбцам по столбцам при помощи axis=1
raitings_dates = pd.concat(
    [raitings, dates_df],
    axis=1
)
display(raitings_dates)
```
