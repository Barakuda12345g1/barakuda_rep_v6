```python
# читаю библиотеку
import pandas as pd


# знакомлюсь с новыми таблицами
raitings1_df = pd.read_csv('data/ratings1.csv', sep=',')
raitings2_df = pd.read_csv('data/ratings2.csv', sep=',')
movies_df = pd.read_csv('data/movies.csv', sep=',')
dates_df = pd.read_csv('data/dates.csv', sep=',')

# выстравляем индексы верно при помощи параметра  ignore_index=True
raitings = pd.concat(
    [raitings1_df, raitings2_df],
    ignore_index=True
)


# убираем дубликаты и правим индекс
raitings.drop_duplicates(ignore_index=True)
# конкатенируем таблицы по столбцам по столбцам при помощи axis=1
raitings_dates = pd.concat(
    [raitings, dates_df],
    axis=1
)
```

## Метод обьединения .join()

### .join()



        other — таблица, которую мы присоединяем. При объединении она является «правой», а исходная таблица, от имени которой вызывается метод, является «левой».
        how — параметр типа объединения. Он может принимать значения 'inner', 'left' (left outer), 'right' (right outer), и 'outer' (full outer). По умолчанию параметр установлен на 'left'.
        on — параметр, который определяет, по какому столбцу в «левой» таблице происходит объединение по индексам из «правой».
        lsuffix и rsuffix — дополнения (суффиксы) к названиям одноимённых столбцов в «левой» и «правой» таблицах.



```python
# тренируемся в присоединении способом 'left outer'
joined_df = raitings_dates.join(
    movies_df,
    how='left',
    rsuffix='_right'
)

display(joined_df.head(15))
display(joined_df)
```

```python
# тренирируемся в присоединении способом 'right outer'
joined_df = raitings_dates.join(
    movies_df,
    how='right',
    lsuffix='_left'
)

display(joined_df.head(10))
display(joined_df.tail(10))
display(movies_df)
```

```python
# создаем присоединение по ключевому признаку
joined_df = raitings_dates.join(
    movies_df.set_index('movieId'),
    on='movieId',
    how='left'
)

display(joined_df.head(15))
display(joined_df.tail(15))
```

## Метод обьединения merge

### .merge()



            right — присоединяемая таблица. По умолчанию она является «правой».
            how — параметр типа объединения. По умолчанию принимает значение 'inner'.
            on — параметр, который определяет, по какому столбцу происходит объединение. Определяется автоматически, но рекомендуется указывать вручную.
            left_on — если названия столбцов в «левой» и «правой» таблицах не совпадают, то данный параметр отвечает за наименования ключевого столбца исходной таблицы.
            right_on — аналогично предыдущему, параметр отвечает за наименование ключевого столбца присоединяемой таблицы.



```python
# произведем слияние более гибким способом в отличии от .join()
merged = raitings_dates.merge(
    movies_df,
    on='movieId',
    how='left'
)
# проверяем, что мы получаем правельно соединенную таблицу
display(merged)
```

```python
print(f'Количестов строк в таблице raitings_dates: {raitings_dates.shape[0]}')
print(f'Количество строк в таблице merged: {merged.shape[0]}')
print(raitings_dates.shape[0] == merged.shape[0])
```

### Особенности использования merge()

```python
merged_full = raitings_dates.merge(
    movies_df,
    on='movieId',
    how='outer'
)
# видим, что количество строк в исходной таблице увеличелось
print(f'Число строк в таблице merged_full: {merged_full.shape[0]}')
display(merged_full.tail())
```

```python
# метод .merge() работатет как при соединение таблицы методом .concat()
# запоминаем, что при параметре how='outer' вторая таблица будет полностью
# соединена с первой и пойдет в конец
# при этом все столбцы должны быть ключевыми т.е одинаковыми
merge_raitings = raitings1_df.merge(raitings2_df, how='outer')
print(f'Число строк в таблице merge_raitings: {merge_raitings.shape[0]}')
display(merge_raitings)
```
