```python
# импортируем библиотеки для работы с данными, математических преобразований
# и для визуализации
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# читаем в переменную датасет для работы с данными
sber_data = pd.read_csv('data/sber_data.csv', sep=',')
display(sber_data.head())
display(sber_data.tail())

# узнаем количество строк
display(sber_data.info())
display(sber_data.shape)

# узнаем количество районов 
display(sber_data['sub_area'].value_counts())
display(sber_data['sub_area'].nunique())

# узнаем максимальное значение в признаке price_doc - цена
display(sber_data['price_doc'].describe())
display(sber_data['price_doc'].max())

# строю кообчатую диаграмму 
# для того чтобы посмотреть зависимость экологической обстановки
# на цену квартиры
fig = plt.figure(figsize=(10, 7))
boxplot = sns.boxplot(data=sber_data, 
                      x='ecology',
                      y='price_doc',
                      )
boxplot.set_title('Зависимость цены от экологии');
boxplot.grid();
boxplot.set_xlabel('Экология');
boxplot.set_ylabel('Цена')

# описываем диаграмму рассеяния для визуализации зависимости цены price_doc
# и расстояния до центра Москвы kremlin_km

fig = plt.figure(figsize=(10, 5))
scatterplot = sns.scatterplot(
    data=sber_data,
    y='price_doc',
    x='kremlin_km'
)
scatterplot.set_title('Взяимосвязь цены и расстояния до центра Москвы');
scatterplot.set_xlabel('Расстояние');
scatterplot.set_ylabel('Цена')
```
