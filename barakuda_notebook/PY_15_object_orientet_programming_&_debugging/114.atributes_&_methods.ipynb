# Атрибуты и методы

```python
# создаю новый класс
class SalesReport():
    pass

# создаю обьект по классу
report_1 = SalesReport()
# добаляю обьекту класса атрибут
report_1.amount = 10
# создаю еще однин обьект по классу и создаю атрибут
report_2 = SalesReport()
report_2.amount = 20
# создаю всомогательную функцию для того, чтобы увилеть 
# как работает атрибуты обьекта
def print_report(report):
    print('Total amount:', report.amount)

# утверждаем, что отображение атрибута обьекта принадлежащего 
# классу SalesReport происходит корректно
print_report(report_1)
print_report(report_2)
```

```python
# добавляю в класс вышепредствленную функцию
class SalesReport():
    # прописываю в аргументации self, и представляю функцию в качествн метода
    def print_report(self):
        print('Total amount:', self.amount)

# создаю обьекты по классу и добаваляю свойства для того, 
# чтобы увидеьть как работает метод
report_1 = SalesReport()
report_1.amount = 10
report_2 = SalesReport()
report_2.amount = 20

report_1.print_report()
report_2.print_report()
```

```python
# создаю лаконичную логику внутри класса, переопределяю класс снова
class SalesReport():
    # созадю метод, с плюс один аргументом
    def add_deal(self, amount):
        # проверяю на наличме атрибута deals,ввожу экзкмпляр класса
        # и название атрибута в виде строки 
        # в качестве аргумета проверяющего меотожда hasattr
        if not hasattr(self, 'deals'):
            self.deals = []
        # назначю атрибут обьеекта класса в нутри класса
        self.deals.append(amount)
        
    # создаю функцию в качестве метода внутри класса 
    # для того чтобы посчитать колчичество сделок
    def total_amount(self):
        # для отображения будет использованно то что возвращаетмсяи из функциии
        return sum(self.deals)
    
    # создаю функцию, которая будет показывать результа суммирования сделок
    def print_report(self):
        # спользуем вышепредставленную функцию в качестве метода класса для 
        # того чтобы показать результат сложения чисел в списке
        print('Total sales:', self.total_amount())
        
# классифицирую обьект
report = SalesReport()        
# добавляем в список сделки
report.add_deal(10000)
report.add_deal(30000)
# утверждаем корректность вывода
report.print_report()
        
```

```python
class SalesReport():
    # использую технический метод для того, чтобы увидеть удобство того, 
    # что можно пропустить метод требующий проверки наличия атрибута
    # и для того чтобы ввести свойство обьекту класса
    def __init__(self):
        self.deals = []
    def add_deals(self, amount):
        self.deals.append(amount)
    def total_amount(self):
        return sum(self.deals)
    def print_report(self):
        print('Total sales:', self.total_amount())

report = SalesReport()
# утверждамем, что ошибки нет
print(report.deals)
print(report.total_amount())
```

```python
class SalesReport():
    # водим аргумент который будет первым при вызове
    def __init__(self, manager_name):
        self.deals = []
        # увеличиваю количество атрибктов
        self.manager_name = manager_name
    def add_deals(self, amount):
        self.deals.append(amount)
    def total_amount(self):
        return sum(self.deals)
    def print_report(self):
        print('Total amount:', self.total_amount())
        print('Manager:', self.manager_name)

# технический метод включается автоматически, а аргумент при вызове бдует вотрым 
# аргументом после self
report = SalesReport('Ivan_Taranov')
report.add_deals(10000)
report.add_deals(30000)
report.print_report()
```

```python
# добвляю в класс еще свойства и атрибуты для того, 
# чтобы улучшить возможности класса
# обьявляю класс
class SalesReport():
    # техническим методом обьявляю атрибуты
    def __init__(self, employee_name):
        self.deals = []
        self.employee_name = employee_name
    # обьявляю метод, который будет пополнять список словарем с 
    # именами компании и сумму сделки
    def add_deal(self, company, amount):
        # добавляю в список словарь
        self.deals.append({'company': company, 'amount': amount})
    # обьявляю метод, который будет счить общую сумму сделок
    def total_amount(self):
        # использую обьект типа генератор и сумирую все сделки
        return sum([deal['amount'] for deal in self.deals])
    # обьявляю метод который будет возвращать сумму сделок в среднем
    def average_deal(self):
        # при помощи вышепердставленого метода нахожу среднее значение
        return self.total_amount() / len(self.deals)
    # инициализирую метод который соберет в список название компаний
    def all_companies(self):
        # при помощи обьекта типа генератор возвращаю множество, чтобы исключить 
        # повторения и помещаю все в список для удобного отображения
        return list(set([deal['company'] for deal in self.deals]))
    # обьявляю метод, который будет печатать информацию
    def print_report(self):
        print('Employee:', self.employee_name)
        print('Total amount:', self.total_amount())
        print('Average amount:', self.average_deal())
        print('Companies:', self.all_companies())
        
# при классификации обьекта в аргументы вводим имя, которое автоматически будет 
# введено в технический метод 
report = SalesReport('Ivan Semenov')

# сохраняю в классе имена компаний и суммы сделок
report.add_deal('PepsiCo', 120000)
report.add_deal('SkyEng', 250000)
report.add_deal('PepsiCo', 20000)

# визуализирую информацию
report.print_report()
```
