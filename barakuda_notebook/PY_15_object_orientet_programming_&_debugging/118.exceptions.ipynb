# try-exept


```python
try:
    *ваш код*
except Ошибка:
    *Код отлова*
else:
    *Код, который выполнится если всё хорошо прошло в блоке try*
finally:
    *Код, который выполнится по любому*
```

```python
# открываем блок, чтобы отловить участок кода в котором предположительно может 
# возникнкуть исключение 
try:
    a = int(input('a: '))
    b = int(input('b: '))
    c = a / b
    print(c)
# завершаем отлов ошибок:
except ZeroDivisionError as e:
    # когда возникнет исключение, программа перейдет в эту часть блока
    print(e)
    print('После иключения')
# после блока отловки исключений программа будет работать далее
print('После После исключений')
    
```

```python
# открываем блок, чтобы отловить участок кода в котором предположительно может 
# возникнкуть исключение 
try:
    a = int(input('a: '))
    b = int(input('b: '))
    c = a / b
    print(c)
# завершаем отлов ошибок:
except ZeroDivisionError as e:
    # когда возникнет исключение, программа перейдет в эту часть блока
    print(e)
    print('После иключения')
# напишем, что программа должна делать если в чати try не возникло ошибки
else:
    print('Ok')
# напишем, что будет происходить после завершения частей try-except-else
finally:
    print('go out')
# после блока отловки исключений программа будет работать далее
print('После После исключений')
    
```

# raise

```python
# узнаем возраст для утверждения умения об обратке исключений
age = int(input('Сколько тебе лет?'))
if age > 100 or age <= 0:
    raise ValueError('Тебе не может быть столько лет')
print(f'Тебе {age} лет.')
```

```python
# узнаем возраст для утверждения умения об обратке исключений
try:
    age = int(input('Сколько тебе лет?'))
    if age > 100 or age <= 0:
        raise ValueError('Тебе не может быть столько лет')
    
    print(f'Тебе {age} лет.')
except ValueError:
    print('Неправельный возраст')
    
```
