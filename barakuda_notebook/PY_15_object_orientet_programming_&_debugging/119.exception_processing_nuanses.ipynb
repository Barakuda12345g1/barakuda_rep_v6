```python
# создаим блок для обработки исключения
try:
    a = 2
    b = 0
    c = a / b
    print(c)
# даем инструкции, если создалось исключение
except Exception:
    # оставляем pass для того, чтобы вернуться позже
    pass
```

```python
# открываем блок отлова исулючений
try:
    # создаю ошибку
    raise ZeroDivisionError
# отлавливаем, срузу несколько исключений
except ArithmeticError:
    print('ArithmeticError')
# сначала отлавливаем потомка
except ZeroDivisionError:
    print('ZeroDivisionError')
```

```python
# открываем блок отлова исключений для правельного отлова ошибок от конкретного 
# класса к более абстрактному
try:
    raise ZeroDivisionError
except ZeroDivisionError:
    print('ZeroDivisionError')
except ArithmeticError:
    print('ArithmeticError')
```

```python
# создаем собственный класс, исключение 
# для отлова ошибок в собственной программе
class MyException(Exception):
    pass

# утверждаем отлов исключений благодаря собственному классу
try:
    raise MyException('message')
except MyException as e:
    # выводим информацию об исключении
    print(e)
```

```python
# создаем класс родитель, исключение, наследуемся от класса Exception
class ParentException(Exception):
    pass

# создаём класс наследник, исключение
class ChildException(ParentException):
    pass
# возбуждаем исключение и выводим информацию о классе родителе
try:
    raise ChildException('message')
except ParentException as e:
    print(e)
```

```python
# создаю класс родителя, наследник Exception
class ParentException(Exception):
    # обьявляю функцию при обьявлении класса
    def __init__(self, message, error):
        # оязателбно создаем выход к родительскому классу Exception, 
        # он будет вызываться после Errors: error
        super().__init__(message)
        # печататем , то что будет при вызове класса ParentException
        print(f'Errors: {error}')

# создаем класс наследник
class ChildException(ParentException):
    # обьявляем функцию, которая сработае при инициализации 
    def __init__(self, message, error):
        # созадем вызов родителського класса 
        super().__init__(message, error)
        
# утверждаем работу классов-исключений, возбуждаем ошибку
try:
    raise ChildException('message', 'error')
except ParentException as e:
    print(e)
    
```
